# MatrixPlot

Demonstrate 2-dimensional matrix plot with OpenTURNS.

```python
import os
import openturns as ot
from othdrplot import MatrixPlot
```

Load the csv data file.

```python
fname = os.path.join(os.path.abspath(''), '../../tests/data/gauss-mixture.csv')
sample = ot.Sample.ImportFromCSVFile(fname)
sample
```

In order to run the algorithm, we must estimate the distribution of the data. In this example, we use a density KernelSmoothing estimator.

```python
mp = MatrixPlot(sample)
```

```python
_ = mp.draw()
```

```python
ks = ot.KernelSmoothing()
distribution = ks.build(sample)
mp = MatrixPlot(sample,distribution)
_ = mp.draw()
```
