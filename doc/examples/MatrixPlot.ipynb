# MatrixPlot

Demonstrate 2-dimensional matrix plot with OpenTURNS.

```python
import os
import openturns as ot
import othdrplot
import openturns.viewer as otv
```

Load the csv data file.

```python
filename = os.path.join(othdrplot.__path__[0], "data", "gauss-mixture.csv")
sample = ot.Sample.ImportFromCSVFile(filename)
sample
```

We can visualize the sample. 

```python
ot.VisualTest_DrawPairs(sample)
```

We can also plot the marginal distribution estimated by kernel smoothing on the diagonal. 

```python
dimension = sample.getDimension()
distribution = ot.ComposedDistribution(
    [ot.KernelSmoothing().build(sample.getMarginal(i)) for i in range(dimension)]
)
grid = ot.VisualTest_DrawPairsMarginals(sample, distribution)
otv.View(grid).getFigure().set_size_inches(6.0, 6.0)
```

The histogram can be used as well.

```python
dimension = sample.getDimension()
distribution = ot.ComposedDistribution(
    [ot.HistogramFactory().build(sample.getMarginal(i)) for i in range(dimension)]
)
grid = ot.VisualTest_DrawPairsMarginals(sample, distribution)
otv.View(grid).getFigure().set_size_inches(6.0, 6.0)
```

## In 3D

We load a 3D sample with 500 points.

```python
filename = os.path.join(othdrplot.__path__[0], "data", "gauss-mixture-3D.csv")
sample = ot.Sample.ImportFromCSVFile(filename)
sample
```

The class also works for 3D samples.

```python
grid = ot.VisualTest_DrawPairs(sample)
view = otv.View(grid).getFigure().set_size_inches(6.0, 6.0)
```

The kernel smoothing estimate can be used on the diagonal.

```python
distribution = ot.ComposedDistribution(
    [ot.KernelSmoothing().build(sample.getMarginal(i)) for i in range(3)]
)
grid = ot.VisualTest_DrawPairsMarginals(sample, distribution)
otv.View(grid).getFigure().set_size_inches(6.0, 6.0)
```
