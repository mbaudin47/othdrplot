# MatrixPlot

Demonstrate 2-dimensional matrix plot with OpenTURNS.

```python
import os
import openturns as ot
from othdrplot import MatrixPlot
```

Load the csv data file.

```python
sample = ot.Sample.ImportFromCSVFile('../../data/gauss-mixture.csv')
sample
```

In order to run the algorithm, we first create the object with the sample.

```python
mp = MatrixPlot(sample)
```

Then we use the `draw()` method. By default, the histogram is on the diagonal.

```python
_ = mp.draw()
```

If the distribution is known or estimated from the data, it can be given to the class, which then draw it in the diagonal.

```python
ks = ot.KernelSmoothing()
distribution = ks.build(sample)
mp = MatrixPlot(sample,distribution)
_ = mp.draw()
```

We load a 3D sample with 500 points.

```python
sample = ot.Sample.ImportFromCSVFile('../../data/gauss-mixture-3D.csv')
sample
```

```python
mp = MatrixPlot(sample)
_ = mp.draw()
```
